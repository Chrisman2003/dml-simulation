```python
import numpy as np
import matplotlib.pyplot as plt
from backend_ols import *
```

```python
# Hyperparameters
mu = 1.0
sigma = 2.0 # Standard deviation of noise
alpha = 0.05
n = 30
n_rep = 1000
```

```python
# Run Monte Carlo simulations
results = monte_carlo(
    n=n,
    mu=mu,
    sigma=sigma,
    alpha=alpha,
    n_rep=n_rep
)

summary = summarize_results(results, mu)
for k, v in summary.items():
    print(f"{k}: {v:.4f}")
```

```python
# Analysis: Central Limit Theorem 
plot_sampling_distribution(results, mu)
```

```python
# Analysis: Confidence intervals plot
plot_confidence_intervals(results, mu)
```

```python
# Analysis: Law of Large Numbers
ns = [10, 30, 100, 300, 1000, 3000]
convergence_experiment(ns=ns, mu=mu, sigma=sigma, alpha=alpha, n_rep=500)
```
